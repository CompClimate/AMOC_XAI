```python
def create_sequences(data, labels, time_steps):
    sequences, sequence_labels = [], []
    for i in range(len(data) - time_steps + 1):
        sequences.append(data[i:i + time_steps])
        sequence_labels.append(labels[i + time_steps - 1])  # Label at the end of the sequence
    return np.array(sequences), np.array(sequence_labels)

time_steps = 16
feature_dim = 4

# Create sequences
X_train_seq, y_train_seq = create_sequences(X_train_scaled, y_train, time_steps)
X_val_seq, y_val_seq = create_sequences(X_val_scaled, y_val, time_steps)
X_test_seq, y_test_seq = create_sequences(X_test_scaled, y_test, time_steps)
input_seq, output_seq = create_sequences(input_scaled, output_data, time_steps)
```

```python
# Proxy for LSTM 
class MLPApproxLSTM(nn.Module):
    def __init__(self):
        super().__init__()
        # Map 4 input features -> 8 “LSTM-like” features
        self.fc0 = nn.Linear(4, 8)
        self.relu0 = nn.ReLU()
        
        # Dense layers copied from Keras
        self.fc1 = nn.Linear(8, 4)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(4, 2)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(2, 1)

    def forward(self, x):
        x = self.relu0(self.fc0(x))
        x = self.relu1(self.fc1(x))
        x = self.relu2(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
import numpy as np
import torch
import torch.nn as nn
import tensorflow as tf

class LSTMWithLRP:
    """
    Custom LSTM implementation with LRP based on ArrasL/LRP_for_LSTM
    Adapted for your architecture: LSTM(8) -> Dense(4) -> Dense(2) -> Dense(1)
    """
    
    def __init__(self, keras_model):
        """Initialize from your Keras model"""
        self.extract_weights(keras_model)
        
    def extract_weights(self, keras_model):
        """Extract weights from Keras model"""
        # LSTM weights (layer 0)
        lstm_weights = keras_model.layers[0].get_weights()
        W_i = lstm_weights[0]  # Input weights (4, 32)
        W_h = lstm_weights[1]  # Hidden weights (8, 32)
        b = lstm_weights[2]    # Bias (32,)
        
        # Split LSTM weights by gate (i, f, c, o)
        self.W_i_i = W_i[:, 0:8]   # input gate - input weights
        self.W_i_f = W_i[:, 8:16]  # forget gate - input weights
        self.W_i_c = W_i[:, 16:24] # cell gate - input weights
        self.W_i_o = W_i[:, 24:32] # output gate - input weights
        
        self.W_h_i = W_h[:, 0:8]   # input gate - hidden weights
        self.W_h_f = W_h[:, 8:16]  # forget gate - hidden weights
        self.W_h_c = W_h[:, 16:24] # cell gate - hidden weights
        self.W_h_o = W_h[:, 24:32] # output gate - hidden weights
        
        self.b_i = b[0:8]    # input gate bias
        self.b_f = b[8:16]   # forget gate bias
        self.b_c = b[16:24]  # cell gate bias
        self.b_o = b[24:32]  # output gate bias
        
        # Dense layer weights (layers 1, 2, 3)
        w1, b1 = keras_model.layers[1].get_weights()  # Dense(4): (8,4)
        w2, b2 = keras_model.layers[2].get_weights()  # Dense(2): (4,2)
        w3, b3 = keras_model.layers[3].get_weights()  # Dense(1): (2,1)
        
        # Don't transpose - Keras weights are already in correct shape for np.dot
        self.W_dense1, self.b_dense1 = w1, b1  # (8,4)
        self.W_dense2, self.b_dense2 = w2, b2  # (4,2)
        self.W_dense3, self.b_dense3 = w3, b3  # (2,1)
        
        # Model dimensions
        self.input_dim = 4
        self.hidden_dim = 8
        self.time_steps = 16
        
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-np.clip(x, -500, 500)))
    
    def tanh(self, x):
        return np.tanh(np.clip(x, -500, 500))
    
    def relu(self, x):
        return np.maximum(0, x)
    
    def forward(self, x):
        """Forward pass through LSTM and dense layers"""
        batch_size = x.shape[0]
        
        # Initialize hidden and cell states
        h = np.zeros((batch_size, self.hidden_dim))
        c = np.zeros((batch_size, self.hidden_dim))
        
        # Store all states for LRP
        self.states = []
        
        # Process each time step
        for t in range(self.time_steps):
            x_t = x[:, t, :]  # Current input
            
            # LSTM gates
            i_t = self.sigmoid(np.dot(x_t, self.W_i_i) + np.dot(h, self.W_h_i) + self.b_i)
            f_t = self.sigmoid(np.dot(x_t, self.W_i_f) + np.dot(h, self.W_h_f) + self.b_f)
            o_t = self.sigmoid(np.dot(x_t, self.W_i_o) + np.dot(h, self.W_h_o) + self.b_o)
            g_t = self.tanh(np.dot(x_t, self.W_i_c) + np.dot(h, self.W_h_c) + self.b_c)
            
            # Update cell and hidden state
            c = f_t * c + i_t * g_t
            h = o_t * self.tanh(c)
            
            # Store states for backpropagation
            self.states.append({
                'x': x_t, 'h_prev': h if t == 0 else self.states[t-1]['h'],
                'c_prev': c if t == 0 else self.states[t-1]['c'],
                'i': i_t, 'f': f_t, 'o': o_t, 'g': g_t,
                'c': c, 'h': h
            })
        
        # Dense layers (using last LSTM output)
        self.lstm_output = h
        self.dense1_input = h
        self.dense1_output = self.relu(np.dot(h, self.W_dense1) + self.b_dense1)
        
        self.dense2_input = self.dense1_output
        self.dense2_output = self.relu(np.dot(self.dense1_output, self.W_dense2) + self.b_dense2)
        
        self.dense3_input = self.dense2_output
        self.output = np.dot(self.dense2_output, self.W_dense3) + self.b_dense3  # Logits
        
        return self.output
    
    def lrp_linear(self, w, x, R, bias=None):
        """LRP for linear layer"""
        w_p = np.maximum(0, w)
        w_n = np.minimum(0, w)
        
        x_p = np.maximum(0, x)
        x_n = np.minimum(0, x)
        
        if bias is not None:
            bias_p = np.maximum(0, bias)
            bias_n = np.minimum(0, bias)
            
            z_p = np.dot(x_p, w_p) + bias_p
            z_n = np.dot(x_n, w_n) + bias_n
        else:
            z_p = np.dot(x_p, w_p)
            z_n = np.dot(x_n, w_n)
        
        z = z_p + z_n
        z += 1e-12 * (z >= 0) - 1e-12 * (z < 0)  # Numerical stability
        
        s_p = R / z * z_p
        s_n = R / z * z_n
        
        c_p = np.dot(s_p, w_p.T)
        c_n = np.dot(s_n, w_n.T)
        
        return c_p + c_n
    
    def lrp_relu(self, x, R):
        """LRP for ReLU activation"""
        return R * (x > 0)
    
    def compute_lrp(self):
        """Compute LRP attributions"""
        # Start from output layer
        R = self.output.copy()
        
        # Dense layer 3 (output)
        R_dense2 = self.lrp_linear(self.W_dense3, self.dense3_input, R, self.b_dense3)
        
        # Dense layer 2 + ReLU
        R_dense2 = self.lrp_relu(self.dense2_output, R_dense2)
        R_dense1 = self.lrp_linear(self.W_dense2, self.dense2_input, R_dense2, self.b_dense2)
        
        # Dense layer 1 + ReLU  
        R_dense1 = self.lrp_relu(self.dense1_output, R_dense1)
        R_lstm = self.lrp_linear(self.W_dense1, self.dense1_input, R_dense1, self.b_dense1)
        
        # LRP through LSTM (simplified - distribute to last timestep)
        # For full temporal attribution, you'd need to implement the complete LSTM LRP
        # This gives attribution to the final hidden state
        
        # Initialize relevance for inputs
        R_inputs = np.zeros((R_lstm.shape[0], self.time_steps, self.input_dim))
        
        # Simplified: attribute final hidden state relevance to last few timesteps
        # More sophisticated implementation would backpropagate through all LSTM computations
        for t in range(max(0, self.time_steps-3), self.time_steps):
            # Distribute relevance based on input contribution to gates
            state = self.states[t]
            x_t = state['x']  # Shape: (1, 4)
            
            # Calculate how much each input feature contributed to the gates at timestep t
            # Use input weights to estimate contribution
            gate_contributions = (
                np.abs(np.dot(x_t, self.W_i_i)) + 
                np.abs(np.dot(x_t, self.W_i_f)) + 
                np.abs(np.dot(x_t, self.W_i_c)) + 
                np.abs(np.dot(x_t, self.W_i_o))
            ) / 4  # Average across gates, shape: (1, 8)
            
            # Project hidden relevance back to input space through averaged weights
            avg_input_weights = (self.W_i_i + self.W_i_f + self.W_i_c + self.W_i_o) / 4  # (4, 8)
            input_relevance = np.dot(R_lstm * gate_contributions, avg_input_weights.T)  # (1, 4)
            
            R_inputs[:, t, :] = input_relevance / 3  # Distribute over last 3 timesteps
        
        return R_inputs

# Usage with your model
def run_lstm_lrp(keras_model, input_sequences):
    """
    Run LRP on your LSTM model
    
    Args:
        keras_model: Your trained Keras LSTM model
        input_sequences: Input sequences (samples, time_steps, features)
    
    Returns:
        attributions: LRP attributions (samples, time_steps, features)
    """
    # Remove sigmoid for logits
    keras_model.layers[-1].activation = tf.keras.activations.linear
    
    # Create LRP model
    lstm_lrp = LSTMWithLRP(keras_model)
    
    # Compute attributions for each sample
    all_attributions = []
    
    for i in range(input_sequences.shape[0]):
        sample = input_sequences[i:i+1]  # Keep batch dimension
        
        # Forward pass
        output = lstm_lrp.forward(sample)
        
        # Compute LRP
        attribution = lstm_lrp.compute_lrp()
        all_attributions.append(attribution)
        
        if i % 100 == 0:
            print(f"Processed {i}/{input_sequences.shape[0]} samples")
    
    return np.concatenate(all_attributions, axis=0)

# Example usage:

keras_model = tf.keras.models.load_model("models/lstm/stationary/ensemble_member_0.keras")
attributions = run_lstm_lrp(keras_model, input_seq)
print(f"Attribution shape: {attributions.shape}")  # (samples, 16, 4)

# Visualize attributions for first sample
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))
plt.imshow(attributions[0].T, aspect='auto', cmap='RdBu_r')
plt.xlabel('Time Steps')
plt.ylabel('Features')
plt.title('LRP Attributions')
plt.colorbar()
plt.show()
```

```python
import numpy as np
from numpy import newaxis as na


def lrp_linear(hin, w, b, hout, Rout, bias_nb_units, eps, bias_factor=0.0, debug=False):
    """
    LRP for a linear layer with input dim D and output dim M.
    Args:
    - hin:            forward pass input, of shape (D,)
    - w:              connection weights, of shape (D, M)
    - b:              biases, of shape (M,)
    - hout:           forward pass output, of shape (M,) (unequal to np.dot(w.T,hin)+b if more than one incoming layer!)
    - Rout:           relevance at layer output, of shape (M,)
    - bias_nb_units:  total number of connected lower-layer units (onto which the bias/stabilizer contribution is redistributed for sanity check)
    - eps:            stabilizer (small positive number)
    - bias_factor:    set to 1.0 to check global relevance conservation, otherwise use 0.0 to ignore bias/stabilizer redistribution (recommended)
    Returns:
    - Rin:            relevance at layer input, of shape (D,)
    """
    sign_out = np.where(hout[na,:]>=0, 1., -1.) # shape (1, M)
    
    numer    = (w * hin[:,na]) + ( bias_factor * (b[na,:]*1. + eps*sign_out*1.) / bias_nb_units ) # shape (D, M)
    # Note: here we multiply the bias_factor with both the bias b and the stabilizer eps since in fact
    # using the term (b[na,:]*1. + eps*sign_out*1.) / bias_nb_units in the numerator is only useful for sanity check
    # (in the initial paper version we were using (bias_factor*b[na,:]*1. + eps*sign_out*1.) / bias_nb_units instead)
    
    denom    = hout[na,:] + (eps*sign_out*1.)   # shape (1, M)
    
    message  = (numer/denom) * Rout[na,:]       # shape (D, M)
    
    Rin      = message.sum(axis=1)              # shape (D,)
    
    if debug:
        print("local diff: ", Rout.sum() - Rin.sum())
    # Note: 
    # - local  layer   relevance conservation if bias_factor==1.0 and bias_nb_units==D (i.e. when only one incoming layer)
    # - global network relevance conservation if bias_factor==1.0 and bias_nb_units set accordingly to the total number of lower-layer connections 
    # -> can be used for sanity check
    
    return Rin
```

```python
class LSTM_LRP:
    
    def __init__(self, i):
        model = tf.keras.models.load_model(f"models/lstm/stationary/ensemble_member_{i}.keras")
        # LSTM layer 
        lstm_layer = model.layers[0]
        # Wxh (x -> h)
        self.Wxh = lstm_layer.get_weights()[0]
        # Whh (h -> h)
        self.Whh = lstm_layer.get_weights()[1]
        # bias 
        self.b = lstm_layer.get_weights()[2]

        # Dense layers 
        self.dense1_W, self.dense1_b = model.layers[1].get_weights()
        self.dense2_W, self.dense2_b = model.layers[2].get_weights()
        self.out_W, self.out_b = model.layers[3].get_weights()
        
        # units
        self.units = lstm_layer.units
    
    def set_input(self, x):
        
        T = len(x)
        self.x = x
        self.T = T
        self.h = np.zeros((T+1, self.units))
        self.c = np.zeros((T+1, self.units))
    
    def forward(self):
        
        # gate indices
        d = self.units
        idx    = np.hstack((np.arange(0,d), np.arange(2*d,4*d))).astype(int)
        idx_i, idx_g, idx_f, idx_o = np.arange(0, d), np.arange(d,2*d), np.arange(2*d,3*d), np.arange(3*d,4*d)
        
        # initialize
        self.gates_xh = np.zeros((self.T, 4*d))
        self.gates_hh = np.zeros((self.T, 4*d))
        self.gates_pre = np.zeros((self.T, 4*d))
        self.gates = np.zeros((self.T, 4*d))
        
        for t in range(self.T):
            # LSTM part
            self.gates_xh[t]     = np.dot(self.x[t], self.Wxh)        
            self.gates_hh[t]     = np.dot(self.h[t-1], self.Whh) 
            self.gates_pre[t]    = self.gates_xh[t] + self.gates_hh[t] + self.b
            
            self.gates[t,idx]    = 1.0/(1.0 + np.exp(-self.gates_pre[t,idx]))
             
            self.gates[t,idx_g]  = np.tanh(self.gates_pre[t,idx_g]) 
            self.c[t]            = self.gates[t,idx_f]*self.c[t-1] + self.gates[t,idx_i]*self.gates[t,idx_g]
            self.h[t]            = self.gates[t,idx_o]*np.tanh(self.c[t])
            
        # Dense part 
        self.a1 = np.maximum(0, np.dot(self.h[self.T-1], self.dense1_W) + self.dense1_b)  # ReLU
        self.a2 = np.maximum(0, np.dot(self.a1, self.dense2_W) + self.dense2_b)           # ReLU
        self.y = np.dot(self.a2, self.out_W) + self.out_b # linear output
        #self.y = 1 / (1 + np.exp(-self.z)) # sigmoid 
        
        return self.y.copy()
    
    def lrp(self, eps=1e-6, bias_factor=0.0):
        
        # sequence length and units
        T = self.T 
        d = self.units
        
        # Gate indices
        idx = np.hstack((np.arange(0,d), np.arange(2*d,4*d)))  # i,f,o
        idx_i, idx_g, idx_f, idx_o = np.arange(0,d), np.arange(d,2*d), np.arange(2*d,3*d), np.arange(3*d,4*d)

        # ---------------------------
        # Step 1: Output relevance
        # ---------------------------
        R_out = self.y.copy()  # linear output 

        # ---------------------------
        # Step 2: Dense layers
        # ---------------------------
        R2 = lrp_linear(self.a2, self.out_W, self.out_b, self.y, R_out, self.a2.size, eps, bias_factor)
        R1 = lrp_linear(self.a1, self.dense2_W, self.dense2_b, self.a2, R2, self.a1.size, eps, bias_factor)
        Rh = lrp_linear(self.h[T-1], self.dense1_W, self.dense1_b, self.a1, R1, self.h[T-1].size, eps, bias_factor)

        # ---------------------------
        # Step 3: LSTM relevance
        # ---------------------------
        Rx = np.zeros_like(self.x)
        Rc = np.zeros((T+1, d))
        Rh_t = np.zeros((T+1, d))
        Rh_t[T-1] = Rh

        for t in reversed(range(T)):
            Rc[t] += Rh_t[t]

            # Input gate relevance
            Rg = lrp_linear(
                self.gates[t, idx_i] * self.gates[t, idx_g],
                np.identity(d),
                np.zeros(d),
                self.c[t],
                Rc[t],
                d+4,
                eps,
                bias_factor
            )

            # Input x relevance (fix: axis 1 of Wxh)
            Rx[t] = lrp_linear(
                self.x[t],
                self.Wxh[:, idx_g],
                self.b[idx_g],
                self.gates_pre[t, idx_g],
                Rg,
                d+4,
                eps,
                bias_factor
            )

            # Previous hidden relevance (fix: axis 1 of Whh)
            Rh_t[t-1] = lrp_linear(
                self.h[t-1],
                self.Whh[:, idx_g],
                self.b[idx_g],
                self.gates_pre[t, idx_g],
                Rg,
                d+4,
                eps,
                bias_factor
            )

            # Previous cell relevance
            Rc[t-1] = lrp_linear(
                self.gates[t, idx_f] * self.c[t-1],
                np.identity(d),
                np.zeros(d),
                self.c[t],
                Rc[t],
                d+4,
                eps,
                bias_factor
            )

        return Rx
```

```python
class LSTM_LRP:
    
    def __init__(self, i):
        model = tf.keras.models.load_model(f"models/lstm/stationary/ensemble_member_{i}.keras")
        # LSTM layer 
        lstm_layer = model.layers[0]
        # Wxh (x -> h)
        self.Wxh = lstm_layer.get_weights()[0]
        # Whh (h -> h)
        self.Whh = lstm_layer.get_weights()[1]
        # bias 
        self.b = lstm_layer.get_weights()[2]

        # Dense layers 
        self.dense1_W, self.dense1_b = model.layers[1].get_weights()
        self.dense2_W, self.dense2_b = model.layers[2].get_weights()
        self.out_W, self.out_b = model.layers[3].get_weights()
        
        # units
        self.units = lstm_layer.units
    
    def set_input(self, x):
        
        T = len(x)
        self.x = x
        self.T = T
        self.h = np.zeros((T+1, self.units))
        self.c = np.zeros((T+1, self.units))
    
    def forward(self):
        
        # gate indices
        d = self.units
        idx    = np.hstack((np.arange(0,d), np.arange(2*d,4*d))).astype(int)
        idx_i, idx_g, idx_f, idx_o = np.arange(0, d), np.arange(d,2*d), np.arange(2*d,3*d), np.arange(3*d,4*d)
        
        # initialize
        self.gates_xh = np.zeros((self.T, 4*d))
        self.gates_hh = np.zeros((self.T, 4*d))
        self.gates_pre = np.zeros((self.T, 4*d))
        self.gates = np.zeros((self.T, 4*d))
        
        for t in range(self.T):
            # LSTM part
            self.gates_xh[t]     = np.dot(self.x[t], self.Wxh)        
            self.gates_hh[t]     = np.dot(self.h[t-1], self.Whh) 
            self.gates_pre[t]    = self.gates_xh[t] + self.gates_hh[t] + self.b
            
            self.gates[t,idx]    = 1.0/(1.0 + np.exp(-self.gates_pre[t,idx]))
             
            self.gates[t,idx_g]  = np.tanh(self.gates_pre[t,idx_g]) 
            self.c[t]            = self.gates[t,idx_f]*self.c[t-1] + self.gates[t,idx_i]*self.gates[t,idx_g]
            self.h[t]            = self.gates[t,idx_o]*np.tanh(self.c[t])
            
        # Dense part 
        self.a1 = np.maximum(0, np.dot(self.h[self.T-1], self.dense1_W) + self.dense1_b)  # ReLU
        self.a2 = np.maximum(0, np.dot(self.a1, self.dense2_W) + self.dense2_b)           # ReLU
        self.y = np.dot(self.a2, self.out_W) + self.out_b # linear output
        #self.y = 1 / (1 + np.exp(-self.z)) # sigmoid 
        
        return self.y.copy()
    
    def lrp(self, x, eps=1e-6, bias_factor=0.0):
        
        # forward pass 
        self.set_input(x)
        self.forward()
        
        # sequence length and units
        T = self.T 
        d = self.units
        e = 4
        
        # Gate indices
        idx = np.hstack((np.arange(0,d), np.arange(2*d,4*d)))  # i,f,o
        idx_i, idx_g, idx_f, idx_o = np.arange(0,d), np.arange(d,2*d), np.arange(2*d,3*d), np.arange(3*d,4*d)
        
        # going backwards so first output
        R_out = self.y.copy()  # linear output 

        # next, dense layers
        R2 = lrp_linear(self.a2, self.out_W, self.out_b, self.y, R_out, self.a2.size, eps, bias_factor)
        R1 = lrp_linear(self.a1, self.dense2_W, self.dense2_b, self.a2, R2, self.a1.size, eps, bias_factor)
        Rh = lrp_linear(self.h[T-1], self.dense1_W, self.dense1_b, self.a1, R1, self.h[T-1].size, eps, bias_factor)

        # lstm layer
        Rx = np.zeros_like(self.x)
        Rc = np.zeros((T+1, d))
        Rg = np.zeros((T, d))
        Rh_t = np.zeros((T+1, d))
        Rh_t[T-1] = Rh

        for t in reversed(range(T)):
            Rc[t] += Rh_t[t]
            Rc[t-1] = lrp_linear(
                self.gates[t, idx_f] * self.c[t-1],
                np.identity(d),
                np.zeros(d),
                self.c[t],
                Rc[t],
                2*d,
                eps,
                bias_factor
            )
            Rg[t] = lrp_linear(
                self.gates[t, idx_i] * self.gates[t, idx_g],
                np.identity(d),
                np.zeros(d),
                self.c[t],
                Rc[t],
                2*d,
                eps,
                bias_factor
            )

            # Input x relevance (fix: axis 1 of Wxh)
            Rx[t] = lrp_linear(
                self.x[t],
                self.Wxh[:, idx_g],
                self.b[idx_g],
                self.gates_pre[t, idx_g],
                Rg[t],
                d+e,
                eps,
                bias_factor
            )

            # Previous hidden relevance (fix: axis 1 of Whh)
            Rh_t[t-1] = lrp_linear(
                self.h[t-1],
                self.Whh[:, idx_g],
                self.b[idx_g],
                self.gates_pre[t, idx_g],
                Rg[t],
                d+e,
                eps,
                bias_factor
            )

        return Rx
```
