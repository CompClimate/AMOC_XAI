```python
import keras
import numpy as np
from keras.models import Sequential, Model
from keras.layers import Dense, Reshape, Flatten, Input, Dropout, LSTM
from keras.optimizers import Adam, SGD, RMSprop
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import os
```

```python
# global constants
Sv=1.e9 # m^3/sec
```

```python
# salinity only, linear forcing 
input_data = np.load('data/input_sal_temp_ns.npy')[:, 1:]
output_data = np.load('data/output_sal_temp_ns.npy')
```

```python
X_train = input_data[0:50000]
y_train = output_data[0:50000]
X_val = input_data[50000:75000]
y_val = output_data[50000:75000]
X_test = input_data[75000:100000]
y_test = output_data[75000:100000] 
```

```python
y_train.shape
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
scaler.mean_,scaler.scale_

X_train_scaled = scaler.transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
input_scaled = scaler.transform(input_data)

Y_train = keras.utils.to_categorical(y_train)
Y_val = keras.utils.to_categorical(y_val)
Y_test = keras.utils.to_categorical(y_test)
output_scaled = keras.utils.to_categorical(output_data)
```

```python
# used ChatGPT here 

import numpy as np

def create_sequences(data, labels, time_steps):
    """
    Create sequences and corresponding labels for LSTM.
    Args:
        data: 2D array of shape (n_samples, n_features)
        labels: 2D array of shape (n_samples, n_classes)
        time_steps: Number of time steps in each sequence
    Returns:
        sequences: 3D array of shape (n_sequences, time_steps, n_features)
        sequence_labels: 2D array of shape (n_sequences, n_classes)
    """
    sequences, sequence_labels = [], []
    for i in range(len(data) - time_steps + 1):
        sequences.append(data[i:i + time_steps])
        sequence_labels.append(labels[i + time_steps - 1])  # Label at the end of the sequence
    return np.array(sequences), np.array(sequence_labels)

# Example input
time_steps = 10
#X = np.random.rand(30000, 5)  # Input data
#y = np.random.randint(0, 2, size=(30000, 2))  # One-hot encoded labels

# Create sequences
X_train_seq, y_train_seq = create_sequences(X_train_scaled, Y_train, time_steps)
X_val_seq, y_val_seq = create_sequences(X_val_scaled, Y_val, time_steps)
X_test_seq, y_test_seq = create_sequences(X_test_scaled, Y_test, time_steps)
print("Input shape:", X_val_seq.shape)  # (29991, 10, 5)
print("Labels shape:", y_train_seq.shape)  # (29991, 2)
```

```python
# training the ensemble 
time_steps = 10  # Number of sequential vectors
feature_dim = 4  # Each vector has 5 features
batch_size = 32  # Batch size
for i in range(10):
    print(i)
    model = Sequential([
    LSTM(8, input_shape=(time_steps, feature_dim), return_sequences=False),  # LSTM layer
    Dense(4, activation='relu'),  # Optional dense layer for additional learning
    Dense(2, activation='relu'),
    Dense(2, activation='softmax')  # Output layer for binary classification
    ])
    model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
    es = keras.callbacks.EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=5)
    history = model.fit(X_train_seq, y_train_seq, validation_data=(X_val_seq, y_val_seq), 
          epochs=10, batch_size=batch_size, callbacks=[es])
    os.makedirs("models_lstm_ensemble", exist_ok=True)
    model.save('models_lstm_ensemble/stommel_lstm_ns_'+str(i)+'.h5')
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

# Example: Input data dimensions
time_steps = 10  # Number of sequential vectors
feature_dim = 4  # Each vector has 5 features
batch_size = 32  # Batch size

# Generate dummy data for demonstration
# Input shape: (number of samples, time_steps, feature_dim)
#X_train = np.random.rand(1000, time_steps, feature_dim)
#y_train = np.random.randint(0, 2, size=(1000, 1))  # Binary labels (0 or 1)

#X_test = np.random.rand(200, time_steps, feature_dim)
#y_test = np.random.randint(0, 2, size=(200, 1))

# Build the LSTM model
model = Sequential([
    LSTM(8, input_shape=(time_steps, feature_dim), return_sequences=False),  # LSTM layer
    Dense(4, activation='relu'),  # Optional dense layer for additional learning
    Dense(2, activation='relu'),
    Dense(2, activation='softmax')  # Output layer for binary classification
])

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model
model_eval = model.fit(X_train_seq, y_train_seq, validation_data=(X_val_seq, y_val_seq), 
          epochs=10, batch_size=batch_size)

# Evaluate the model
loss, accuracy = model.evaluate(X_test_seq, y_test_seq)
print(f"Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}")
```

```python
plt.plot(model_eval.history['loss'], label='Training Loss');
plt.plot(model_eval.history['val_loss'], label='Validation Loss');
plt.title('Loss Plot');
plt.legend();
```

```python
import numpy as np
from tensorflow.keras.models import load_model

# Prepare storage for predictions from each ensemble model
datapredicted_results_ensemble = np.zeros((10, input_seq.shape[0], 2))  # Remove np.nan initialization

for i in range(10):
    print(f"Loading and predicting with model {i + 1}...")
    
    # Load the trained model directly
    model_path = f'models_lstm_ensemble/stommel_lstm_ns_{i}.h5'
    model = load_model(model_path)
    
    # Predict with the model
    datapredicted = model.predict(input_seq)
    
    # Store predictions in the ensemble array
    datapredicted_results_ensemble[i, ...] = datapredicted

# Aggregate ensemble predictions (e.g., averaging)
final_predictions = np.mean(datapredicted_results_ensemble, axis=0)

# Optionally, get the class with the highest probability
final_class_predictions = np.argmax(final_predictions, axis=1)
```
