```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# set parameters:
meter=1;
year=365*24*3600;
S0=35.0;
T0=24.0
DeltaT=-10;
time_max=150000*year;
alpha=0.2; # (kg/m^3) K^-1
beta=0.8; # (kg/m^3) ppt^-1
k=10e9 # (m^3/s)/(kg/m^3) #  3e-8; # sec^-1
# area of each of the ocean boxes, rediculous value, needed to get
# all other results at the right order of magnitude:
area=(50000.0e3)**2
depth=4000
V=area*depth # volume of each of the ocean boxes
Sv=1.e9 # m^3/sec
```

```python
def q(DeltaT,DeltaS):
    # THC transport in m^3/sec as function of temperature and salinity
    flow= -k*(-alpha*DeltaT + beta*DeltaS)
    return flow

def Fs_func(time,time_max,is_Fs_time_dependent=True):
    ########################################################################
    # total surface salt flux into northern box

    ########################################################################
    # Specify maximum and minimum of freshwater forcing range during the
    # run in m/year:
    #FW_min=-0.1;
    FW_min=-5;
    #FW_max=5;
    FW_max=5;
    ########################################################################
    if is_Fs_time_dependent:
            #flux=FW_min+(FW_max-FW_min)*time/time_max;
            flux=(FW_max-FW_min)*np.sin(6*np.pi*time/time_max)
    else:
        flux=2;
        
    # convert to total salt flux:
    return flux*area*S0/year

def Ft_func(time, time_max):
    Ft_min = -10
    Ft_max = 10
    flux=(Ft_max-Ft_min)*np.sin(6*np.pi*time/time_max)
    return flux*area*T0/year

def Fs_sin(time):
    return np.sin(2*time)*area*S0/year

Fs_to_m_per_year=S0*area/year

Ft_to_m_per_year=T0*area/year
```

```python
# differential equation for change in salinity
def DeltaS_de(time, DeltaS, time_max, DeltaT):
    dDeltaSdt = 2*np.abs(q(DeltaT, DeltaS))*DeltaS + 2*Fs_func(time, time_max)
    return -dDeltaSdt/V
```

```python
def DeltaT_de(time, DeltaS, time_max, DeltaT):
    dDeltaTdt = 2*np.abs(q(DeltaT, DeltaS))*DeltaT + 2*Ft_func(time, time_max)
    return -dDeltaTdt/V
```

```python
def system_de(time, DeltaS, time_max, DeltaT):
    return [DeltaS_de(time, DeltaS, time_max, DeltaT), DeltaT_de(time, DeltaS, time_max, DeltaT)]
```

```python
def system_de_2(time, S1, S2, time_max):
    q_eval = np.abs(q(0.0, S1-S2))
    dS1dt = (q_eval*(S2-S1) - Fs_func(time, time_max))/V
    dS2dt = (q_eval*(S1-S2) + Fs_func(time, time_max))/V
    return [dS1dt, dS2dt]
```

```python
def system_de_4(time, S1, S2, T1, T2, time_max):
    q_eval = np.abs(q(T1-T2, S1-S2))
    dS1dt = (q_eval*(S2-S1) - Fs_func(time, time_max))/V
    dS2dt = (q_eval*(S1-S2) + Fs_func(time, time_max))/V
    dT1dt = (q_eval*(T2-T1) - Ft_func(time, time_max))/V
    dT2dt = (q_eval*(T1-T2) + Ft_func(time, time_max))/V
    return [dS1dt, dS2dt, dT1dt, dT2dt]
```
