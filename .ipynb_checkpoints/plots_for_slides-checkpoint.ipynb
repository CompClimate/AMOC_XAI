```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_data = np.load('data/input_sal_temp_ns.npy')
```

```python
time = input_data[:, 0]
salinity = input_data[:, 1]
temp = input_data[:, 2]
```

```python
fig, ax = plt.subplots(1, 2, figsize = (20, 8))
ax[0].plot(time, salinity)
ax[0].set_ylabel('Salinity, $\Delta S$ (in ppt)');
ax[0].set_xlabel('Time (in kyr)');
ax[1].set_ylabel('Temperature, $\Delta T$ (in $\deg$ C)');
ax[1].set_xlabel('Time (in kyr)');
```
